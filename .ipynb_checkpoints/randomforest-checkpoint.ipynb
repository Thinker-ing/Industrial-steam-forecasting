```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split 
from statsmodels.stats.outliers_influence import variance_inflation_factor 
import warnings
warnings.filterwarnings("ignore")
```

```python
#读取数据
train_data0=pd.read_csv('../data/zhengqi_train.txt', sep = '\t')
test_data0=pd.read_csv('../data/zhengqi_test.txt', sep = '\t')
```

```python
data = train_data0.append(test_data0,ignore_index = True )
data.info()
```

```python
#训练数据总览
train_data0.info()
```

```python
# 没有空缺值，由于指标的含义暂不清晰，异常值不进行处理
train_data0.describe()
```

```python
#查看target分布情况，在具体分析，可以看到数据是相对平滑的，且没有出现拖尾现象，所以此处暂不做截尾处理。
sns.distplot(train_data0['target'])
plt.show()
```

```python
#训练数据分布情况
#随机游走，布朗运动
for col in test_data0.columns:
    sns.distplot(train_data0[col])
    sns.distplot(test_data0[col])
    plt.show()
```

```python
#通过对比，发现'V5','V9','V11', 'V14', 'V17', 'V19', 'V21', 'V22'六个特征在训练集和测试集的分布相差较大，说明这些特征与结果的相关性较差，
#应该剔除
train_data0.drop(['V5','V9','V11', 'V14', 'V17', 'V19', 'V21', 'V22'], axis = 1, inplace = True)
test_data0.drop(['V5','V9','V11', 'V14', 'V17', 'V19', 'V21', 'V22'], axis = 1, inplace = True)
```

```python
#特征相关性
plt.figure(figsize=(20, 16))
colnm = train_data0.columns.tolist()
mcorr = train_data0[colnm].corr(method="spearman")
mask = np.zeros_like(mcorr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
g = sns.heatmap(mcorr, mask=mask, cmap=cmap, square=True, annot=True, fmt='0.2f')
plt.show()
```

```python
mcorr=mcorr.abs()
numerical_corr=mcorr[mcorr['target']>0.1]['target']
print(numerical_corr.sort_values(ascending=False))
index0 = numerical_corr.sort_values(ascending=False).index
print(train_data0[index0].corr('spearman'))
```

```python
# 自变量与目标变量相关性绝对值0.1以上的有：
# V0,V1,V2,V3,V4,V5,V6,V7,V8,V10,V11,V12,V13,V15,V16,V18,V19,V20,V22,V23,V24,V27,V29,V30,V31,V36,V37
# 自变量相关性0.8以上的有：
# V0: V1 V8 ;
# V1: V8 V27 V31;
# V4: V12;
# V6: V7;
# V8: V27 V31 ;
# V10: V36;
# V15: V29;
# V23: V35;
# 于是，选取特征自变量与目标变量相关性绝对值0.1以上并且剔除高度相关性的自变量。
train_data1=train_data0[['V1','V2','V3','V4','V6','V10','V13','V15',
'V16','V18','V20','V23','V24','V30','V37','target']]
train_data1.head()
```

```python
#多重共线性，<10<100 方差膨胀指数 VIF值越接近于1，多重共线性越轻 
#多重共线性是指自变量之间存在线性相关关系，即一个自变量可以是其他一个或几个自变量的线性组合
new_numerical=['V1','V2','V3','V4','V6','V10','V13','V15',
'V16','V18','V20','V23','V24','V30','V37']
X=np.matrix(train_data1[new_numerical])
VIF_list=[variance_inflation_factor(X, i) for i in range(X.shape[1])]
VIF_list
```

```python
# 多重共线性不明显，暂不需要进一步降维处理。为了降低个别特征较大波动，有可能造成不同特征权重系数变化过大，将数据进行z-score标准化。

train0=train_data1.iloc[:,0:-1]
test0=test_data0[['V1','V2','V3','V4','V6','V10','V13','V15',
'V16','V18','V20','V23','V24','V30','V37']]
target=train_data1.iloc[:,-1]
train=(train0-np.mean(train0,axis=0))/np.std(train0,axis=0)
test=(test0-np.mean(test0,axis=0))/np.std(test0,axis=0)
```

```python
# 预测模型探索

train_data,test_data,train_target,test_target=train_test_split(train,target,test_size=0.2,random_state=1)
m=RandomForestRegressor()
m.fit(train_data, train_target)
score=mean_squared_error(test_target,m.predict(test_data))
print(score)
```

```python
from xgboost import XGBRegressor
m=XGBRegressor()
m.fit(train_data, train_target)
score=mean_squared_error(test_target,m.predict(test_data))
print(score)
```

```python
# 多次重复，查看模型预测结果的稳定性

model_accuracies = []

for repetition in range(100):
    train_data,test_data,train_target,test_target=train_test_split(train,target,test_size=0.2,random_state=1)

    m=RandomForestRegressor()
    m.fit(train_data, train_target)
    score=mean_squared_error(test_target,m.predict(test_data))
    model_accuracies.append(score)

sns.distplot(model_accuracies)
```

```python
model_accuracies = []

for repetition in range(100):
    train_data,test_data,train_target,test_target=train_test_split(train,target,test_size=0.2,random_state=1)
    m=XGBRegressor()
    m.fit(train_data, train_target)
    score=mean_squared_error(test_target,m.predict(test_data))
    model_accuracies.append(score)
sns.distplot(model_accuracies)
```

```python
# 模型参数调优

param_test = {'n_estimators':[90,100,110,120,130],
             'max_depth':[2,3,4,5,6],
             'colsample_bytree':[0.9],
             'eta':[0.08,0.09,0.10,0.11,0.12,0.13]}
# 'max_features':('auto','sqrt','log2')
# param_test2 = {'n_estimators':[195,200],
# 'max_depth':[14,15,16,17],
#            }
# m = GridSearchCV(estimator = RandomForestRegressor(max_features='log2'),  
#    param_grid = param_test2,
#                 cv=5) 

m = GridSearchCV(estimator = XGBRegressor(),  
   param_grid = param_test)
m=m.fit(train_data,train_target)
score=mean_squared_error(test_target,m.predict(test_data))
print(score)
print(m.best_score_)
print(m.best_params_)
```

```python
m = XGBRegressor(n_estimators=110,colsample_bytree=0.9,eta=0.09,max_depth=4)
m.fit(train_data, train_target)
predict = m.predict(test)
np.savetxt('predict.txt',predict)
print('预测并保存成功')
```
